#### 1.Crear una script en Python que permita generar varios anagramas a partir de una palabra con una longitud no mayor a 12 letras.

```python
nodos = {
        '1': {'2': 10, '3': 55, '4': 25, '5': 45},
        '2': {'1': 10, '3': 20, '4': 25, '5': 40},
        '3': {'1': 55, '2': 20, '4': 15, '5': 30},
        '4': {'1': 25, '2': 25, '3': 15, '5': 50},
        '5': {'1': 45, '2': 40, '3': 30, '4': 50}
    } 
    # Dicionario que contiene todos los nodos y costos de moverse a otro nodo
r = [] # Lista que contendra a las rutas generadas
```

```python
def s_rutas(nodo, nodos, ruta, costo):
    ruta.append(nodo)    # Agrego el punto de inicio a la lista
    if len(ruta) > 1: # Realizo la suma del coste del la ruta desde la corriente hasta el último nodo
        costo += nodos[ruta[-2]][nodo]
    if (len(nodos) == len(ruta)) and ((ruta[0]) in nodos[ruta[-1]]): # Si mi ruta contiene todos los nodos y no es un callejón sin salida
        global r # r es mi lista donde guardare todas las rutas generadas
        print (ruta, costo) # muestro la ruta y el coste de la ruta
        r.append([costo,ruta]) # Lo agrego a la lista
        return
    for n in nodos: # Rutas para todas las ciudades posibles aún no utilizadas
        if (n not in ruta):
            s_rutas(n, dict(nodos), list(ruta), costo) # Vuelvo a llamar a mi funcion

            
def factorial(n,i=0):
    if (i<n):           # Caso general
        f = factorial(n,i+1)
        if i==0:
            i=1
            
        resultado = i * f
    elif(n==0):
        resultado = 1
    else:              # Caso Base
        resultado = n
        
    return resultado
```

```python
x=len(nodos) #x es igual al  numero de nodos

print ("Las rutas que se espera generar son:", factorial(x))

"""""Mi funcion la hice inicialmente para todos los nodos como inicio pero se me complicaba más,
    entonces hice que la funcion solo calcule con un inicio, por ello uso un ciclo para calcular con todos los inicios"""""

for i in range(5): #Ciclo para realizar rutas con todos los inicios
    i+=1 #Para no usar otro auxiliar contador y solo usar una variable
    print ("Inicio en:",i)# Mensaje que muestra donde se esta iniciando la ruta
    s_rutas(str(i), nodos, [], 0) #Llamo a mi funcion
    print ("\n") # Salto de linea
    
r.sort() # Manejo de listas, la ordeno de menor a mayor
print ("la ruta más barata es: ", r[0]) # Imprimo la ruta
r.sort(reverse=True) #Manejo de listas, la ordeno de mayor a menor
print ("la ruta más cara es: ", r[0]) # Imprimo la ruta
```
